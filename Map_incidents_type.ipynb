```python
incident_codes_map = pd.read_csv('/content/Toronto_Fire_Incident_Codes.csv')
incident_codes_map.head()
```

```python
# prompt: I want to create a new column in the dataset df such that if there is a mapping of Initial_CAD_Event_Type  in incident_codes_map dataframe then map it otherwise keep it as it is.

# Create a dictionary mapping from 'Incident Code' to 'Incident Type Description'
incident_codes_dict = dict(zip(incident_codes_map['Code'], incident_codes_map['Description']))

# Use the dictionary to map values in 'Initial_CAD_Event_Type'
df['Mapped_Initial_CAD_Event_Type'] = df['Initial_CAD_Event_Type'].map(incident_codes_dict).fillna(df['Initial_CAD_Event_Type'])
```
