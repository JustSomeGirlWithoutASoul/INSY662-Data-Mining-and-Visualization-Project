```python
imp_features = ['Area_of_Origin', 'Building_Status', 'Business_Impact', 'Status_of_Fire_On_Arrival','Incident_Station_Area','Initial_CAD_Event_Type','']

# Set up for multiple plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
axes = axes.flatten()  # Flatten axes array for easy iteration

for i, col in enumerate(imp_features):
    # Group by the feature and count missing values in 'Estimated_Dollar_Loss'
    missing_counts = df.groupby(col)['Estimated_Dollar_Loss'].apply(lambda x: x.isna().sum()
    
    # Filter out groups with zero missing values
    missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)
    
    # Plot only if there are any missing values in this feature's groups
    if not missing_counts.empty:
        # Bar plot for each feature
        ax = axes[i]
        missing_counts.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')
        ax.set_title(f"Missing Values in 'Estimated_Dollar_Loss' by '{col}' Category")
        ax.set_xlabel(col)
        ax.set_ylabel("Count of Missing Values")
        ax.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```
